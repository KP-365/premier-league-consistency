# 1 Introduction

```python
import pandas as pd
import numpy as np
```

```python
goals = pd.read_csv("stats_data/goals.csv")
matches = pd.read_csv("stats_data/matches.csv")
appearances = pd.read_csv("stats_data/appearances.csv")
```

```python
# GOALS TABLE
# Count goals scored by each player

player_goals = (
    goals.groupby('scorer')
    .size()
    .reset_index(name='goals')
)

# APPEARANCES TABLE
# Total minutes + games played

player_appearances = (
    appearances.groupby('player')
    .agg(games=('player', 'count'),minutes=('minutes_played', 'sum')).reset_index())

# MATCHES TABLE

matches['total_match_goals'] = (
    matches['home_score'] + matches['away_score']
)

# MERGE GOALS + APPEARANCES

player_stats = player_goals.merge(
    player_appearances,
    left_on='scorer',
    right_on='player',
    how='left'
)

# GOALS PER 90

player_stats['g90'] = (player_stats['goals'] / player_stats['minutes']) * 90

# TOP SCORERS
top_scorers = (player_stats[['scorer', 'goals', 'games', 'minutes', 'g90']].sort_values(by='goals', ascending=False))

top_scorers_df = pd.DataFrame(top_scorers)
top_scorers_df.head(20)

top_20 = top_scorers_df.head(20)
```

```python
# COUNT GOALS PER PLAYER PER MATCH

goals_per_match = (
    goals.groupby(['event_id', 'scorer'])
    .size()
    .reset_index(name='goals')
)

# MERGE WITH APPEARANCES

df = appearances.merge(
    goals_per_match,
    left_on=['event_id', 'player'],
    right_on=['event_id', 'scorer'],
    how='left'
)

# Replace NaN goals with 0
df['goals'] = df['goals'].fillna(0)

# GOALS PER 90

df['g90'] = (
    df['goals'] / df['minutes_played']
) * 90

# Remove players with 0 minutes
df = df[df['minutes_played'] > 0]

# PLAYER VARIANCE


player_variance = (
    df.groupby('player')['g90']
    .agg(['mean', 'var', 'std', 'count'])
    .reset_index()
)
```

```python
# COMBINE TOP SCORERS WITH VARIANCE

final_stats = top_scorers_df.merge(
    player_variance,
    left_on='scorer',
    right_on='player',
    how='left'
)

final_stats = final_stats.drop(columns=["player", "mean","count"])

final_stats.head(30)
```

**How many points did the players contribute to?**

Now we need to work out how much impact each player had in regards to their goal output to their team. Mehodology includes:
* Calculating the total points for each team
* Calculating the points where the player scored
* Calculating the weight each player has had on the points

```python
# FINDING OUT HOW MANY POINTS EACH TEAM EARNED

def scorer_team_points(row):
    if row['team'] == row['home_team_x']:
        if row['home_score_x'] > row['away_score_x']:
            return 3
        elif row['home_score_x'] == row['away_score_x']:
            return 1
        else:
            return 0

    elif row['team'] == row['away_team_x']:
        if row['away_score_x'] > row['home_score_x']:
            return 3
        elif row['away_score_x'] == row['home_score_x']:
            return 1
        else:
            return 0

    return None
```

```python
# Merge goals with matches to get team and score info

goal_impact = goals.merge(
    matches[['event_id', 'home_team', 'away_team', 'home_score', 'away_score']],
    on='event_id',
    how='left'
)

# Merge team data from appearances
goal_impact = goal_impact.merge(
    appearances[['event_id', 'player', 'team']],
    left_on=['event_id', 'scorer'],
    right_on=['event_id', 'player'],
    how='left'
)
```

```python
# Calculate points earned in games where player scored
goal_impact['points_when_scored'] = goal_impact.apply(
    scorer_team_points,
    axis=1
)
```

```python
# Home team points
matches['home_points'] = matches.apply(
    lambda row: 3 if row['home_score'] > row['away_score']
    else 1 if row['home_score'] == row['away_score']
    else 0,
    axis=1
)

# Away team points
matches['away_points'] = matches.apply(
    lambda row: 3 if row['away_score'] > row['home_score']
    else 1 if row['away_score'] == row['home_score']
    else 0,
    axis=1
)

# Home total points
home_totals = (
    matches.groupby('home_team')['home_points']
    .sum()
    .reset_index()
)

home_totals.columns = ['team', 'points']

# Away totals points
away_totals = (
    matches.groupby('away_team')['away_points']
    .sum()
    .reset_index()
)

away_totals.columns = ['team', 'points']

# Combine to get total points per team
team_points = pd.concat([
    home_totals,
    away_totals
])

# Total points per team
team_points = (
    team_points.groupby('team')['points']
    .sum()
    .reset_index()
)

team_points = team_points.sort_values(by='points',ascending = False).reset_index()
team_points
```

Now that we have the points of each team we can normalise their goal output by dividing the number of points their team recieved when they scored by the total number of points the team recieved.

```python
# ANALYSING PLAYER IMPACT

player_score_impact = (
    goal_impact.groupby('scorer')
    .agg(
        goals=('scorer', 'count'),
        points_when_scored=('points_when_scored', 'sum'),
        scoring_matches=('event_id', 'nunique')
    )
    .reset_index()
)

# Calculate weighted impact (points per scoring match)

player_score_impact['goal_weighted_impact'] = (
    player_score_impact['points_when_scored'] /
    player_score_impact['scoring_matches']
)

# Merge player teams
player_teams = (
    appearances[['player', 'team']]
    .drop_duplicates())

# Merge player teams with score impact
player_score_impact = player_score_impact.merge(
    player_teams,
    left_on='scorer',
    right_on='player',
    how='left'
)

# Merge team points to get total points for each player's team
player_score_impact = player_score_impact.merge(
    team_points,
    on='team',
    how='left'
)

# Normalise impact by total team points to get relative impact
player_score_impact['normalised_impact'] = (
    player_score_impact['points_when_scored'] /
    player_score_impact['points']
)

# Sort by points when scored to see which players had the biggest impact in terms of points earned for their team
player_score_impact = player_score_impact.sort_values(
    by='points_when_scored',
    ascending=False
).reset_index()

# Merge final stats to get variance info for top scorers
player_score_impact = player_score_impact.merge(
    final_stats,
    on='scorer',
    how='left')

player_score_impact
```

Since the Spearman test indicated a strong association between the two variables, an OLS regression model was then used to further investigate the relationship and examine whether player variance could predict changes in normalised_impact.

```python
import statsmodels.api as sm

X = infered_stats['var']
y = infered_stats['normalised_impact']

# Add constant/intercept
X = sm.add_constant(X)

# Fit model
model = sm.OLS(y, X).fit()

# Regression table
print(model.summary())
```

The OLS regression model further supported the Spearman correlation findings by showing that player variance was a statistically significant normalised_impact. The model produced an R2 value of 0.424, suggesting that approximately 42.4% of normalised_impact could be explained by the variance alone. The regression coefficient was positive being 0.9097, indicating that increases in player variance was associated with the increases in normalised_impact.
